```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load audio file
audio_path = "High/1002_DFA_ANG_XX.wav"
y, sr = librosa.load(audio_path, sr=None)

# Compute spectrogram
spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
print(spectrogram)

# Visualize spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(spectrogram, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.show()
```

```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Create a new folder for spectrogram images if it doesn't exist
output_folder = 'Low_Spectrogram'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate through audio files in the 'High' folder
for i in os.listdir("low"):
    # Load audio file
    audio_path = os.path.join("low", i)
    y, sr = librosa.load(audio_path, sr=None)
    
    # Compute spectrogram
    spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    
    # Visualize spectrogram
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(librosa.power_to_db(spectrogram, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
    plt.title('Mel Spectrogram')
    plt.colorbar(format='%+2.0f dB')  # Add color bar if needed
    plt.tight_layout()
    
    # Save spectrogram as image in the new folder
    output_path = os.path.join(output_folder, 'Low_spectrogram_' + str(i) + '.png')
    plt.savefig(output_path, bbox_inches='tight', pad_inches=0.0)
    plt.close()
```

```python
import os 
import numpy as np
import pandas as pd
import cv2
images = []
import matplotlib.pyplot as plt
```

```python
for i in os.listdir("High_Spectrogram"):
    img = cv2.imread("High_Spectrogram/"+i)
    images.append(img)
```

```python
for i in os.listdir("Low_Spectrogram"):
    img = cv2.imread("Low_Spectrogram/"+i)
    images.append(img)
```

```python
len(images)
```

```python
Y = [1]*1122
```

```python
Y  = Y + ([0]*1129)
Y = np.array(Y)
```

```python
X =  np.array(images)
X = X/255.0
```

```python
resized_images = [cv2.resize(img, (128, 128)) for img in X]
X_resized = np.array(resized_images)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resized, Y, test_size=0.1, random_state=42)
```

```python
X_train.shape
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.utils import to_categorical

model = Sequential()

# Add convolutional layers
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.5))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.5))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.5))
# Flatten the output before passing it to the dense layers
model.add(Flatten())

# Add dense layers
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='sigmoid'))  # Change num_classes to the number of classes in your problem

# Compile the model with sparse categorical crossentropy
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
accuracy = accuracy_score(y_test, y_pred_classes)
print(f"Test Accuracy: {accuracy}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.utils import to_categorical

model = Sequential()

# Add convolutional layers
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.5))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.5))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.5))
# Flatten the output before passing it to the dense layers
model.add(Flatten())

# Add dense layers
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='sigmoid'))  # Change num_classes to the number of classes in your problem

# Compile the model with sparse categorical crossentropy
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=35, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
accuracy = accuracy_score(y_test, y_pred_classes)
print(f"Test Accuracy: {accuracy}")
```
