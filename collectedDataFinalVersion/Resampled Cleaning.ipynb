```python
import os


def cleanResampled(foldername):
    for file in os.listdir(foldername):
        df = pd.read_csv(foldername+'/'+file)
        print(file)
        df = df[['time', 'wx', 'wy', 'wz', 'ax', 'ay', 'az']]
        df = df.dropna()
        df.to_csv(foldername+'/'+file, index = False)

        
        
```
