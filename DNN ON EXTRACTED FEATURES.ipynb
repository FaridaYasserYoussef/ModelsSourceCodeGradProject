```python
import pandas as pd
```

```python
df = pd.read_csv("vtdata.csv")
df
```

```python
correlation_matrix = df.corr()

target_correlation = correlation_matrix["label"]
target_correlation
```

```python
df.drop(['mfccs_var', 'mfccs_min', 'mfccs_max', 'mfccs_median'], axis =  1,  inplace = True)
```

```python
df.drop('mfccs_mean', axis =  1,  inplace = True)
```

```python
df.columns
```

```python
X = df[['mean', 'std', 'maxv', 'minv', 'median', 'skew', 'kurt', 'q1', 'q3',
       'mode', 'iqr', 'rmse', 'energy', 'zero_crossings', 'tempo',
       'spec_centroid', 'spectral_bandwidth', 'spectral_contrast',
       'spectral_flatness', 'spectral_rolloff']]
```

```python
Y = df["label"]
```

```python
from sklearn.model_selection import train_test_split
from keras import regularizers
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Sample data

# Define the model
model = Sequential()

# Add layers to the model
model.add(Dense(256, input_dim=20, activation='relu', kernel_regularizer=regularizers.l2(0.01)))  # Input layer with 64 neurons and ReLU activation
model.add(Dense(128,  activation='relu', kernel_regularizer=regularizers.l2(0.01)))  # Input layer with 64 neurons and ReLU activation
model.add(Dense(64,  activation='relu', kernel_regularizer=regularizers.l2(0.01)))  # Input layer with 64 neurons and ReLU activation
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))  # Hidden layer with 32 neurons and ReLU activation
model.add(Dense(2, activation='sigmoid'))  # Output layer with 1 neuron and sigmoid activation for binary classification

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)  # Training with 10 epochs, batch size of 32, and 20% validation split
```

```python
model.save("vparameters.keras")
```
