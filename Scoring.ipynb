```python
# Importing necessary libraries from PyCaret
from pycaret.regression import setup, compare_models

# Import pandas for data manipulation
import pandas as pd

# Load your dataset
data = pd.read_csv('/content/Updated_Driving_Behavior_Data.csv')

# Printing the shape of the dataset to confirm the number of samples
print(data.shape)

# Check if the dataset is too small for default 10 folds and adjust if necessary
number_of_folds = min(7, len(data) - 1)  # Using 5 or one less than number of samples, whichever is smaller

# Initialize the PyCaret environment
# Adjusting fold to the calculated number of folds
setup_data = setup(data=data, target='AverageScore', fold_strategy='kfold', fold=number_of_folds, session_id=123)

# Compare a limited set of models to avoid overfitting on a small dataset
# Here only decision tree and ridge regression models are included
best_model = compare_models()

# Print the best model
print(best_model)
```

```python
from pycaret.regression import create_model, tune_model,predict_model


lr = create_model('lr')


tuned_lr = tune_model(lr)

# Print the tuned model
print(tuned_lr)
```
